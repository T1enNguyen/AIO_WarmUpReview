# **1. Basic function**

## 1.1 Built in Functions

Python có hàng trăm hàm built-in sẵn (không cần định nghĩa).

Ví dụ phổ biến:
- `print()`, `len()`, `type()`, `sum()`, `max()`, `min()`, `sorted()`, `range()`, `input()`

Python built-in functions có thể xem thêm tại [python documentation](https://docs.python.org/3.15/library/functions.html).

```python
# Một số built-in functions phổ biến
print(abs(-7))          # Trả về giá trị tuyệt đối của một số
print(pow(3, 4))       # Tính lũy thừa: 3^4
print(round(3.6))      # Làm tròn số thập phân
print(len("Hello"))   # Đếm số ký tự trong chuỗi
print(max(1, 5, 3))    # Tìm giá trị lớn nhất
print(min(1, 5, 3))    # Tìm giá trị nhỏ nhất
print(sum([1, 2, 3, 4])) # Tính tổng các phần tử trong danh sách
```

```python
name = input("Nhập tên của bạn: ")
print("Xin chào, " + name + "!")
```

```python
numbers = [10, 20, 30, 40, 50]
print(sorted(numbers))  # Sắp xếp danh sách và trả về danh sách mới
```

## 1.2 User-defined Functions

Function là một khối mã hoặc câu lệnh lập trình có thể tái sử dụng, được thiết kế để thực hiện một nhiệm vụ nhất định.

```python
# Định nghĩa hàm user-defined function/Khai báo hàm/Declaring a function
def greet():
    print("Chào mừng bạn đến với Python!")
    
# Gọi hàm/Thực thi hàm/Calling the function
greet()
```

**Hàm không có tham số**

```python
def generate_full_name ():
    first_name = 'Tien'
    last_name = 'Nguyen'
    space = ' '
    full_name = first_name + space + last_name
    print(full_name)
generate_full_name () # calling a function

def add_two_numbers ():
    num_one = 2
    num_two = 3
    total = num_one + num_two
    print(total)
add_two_numbers()
```

**Hàm trả về một giá trị**

Nếu không có câu lệnh `return` nó sẽ trả về giá trị `None`.

```python
def generate_full_name():
    first_name = 'Tien'
    last_name = 'Nguyen'
    space = ' '
    full_name = first_name + space + last_name
    return full_name
print(generate_full_name())

def add_two_numbers():
    num_one = 2
    num_two = 3
    total = num_one + num_two
    return total
print(add_two_numbers())
```

**Hàm với các tham số**

Trong một hàm, chúng ta có thể truyền các kiểu dữ liệu khác nhau (number, string, boolean, list, tuple, dictionary or set) làm tham số.

```python
# Single parameter: Nếu hàm nhận một tham số (a single parameter) thì nên gọi hàm với một đối số (a single argument)
def greetings (name):
    message = name + ', welcome to Python for Everyone!'
    return message

print(greetings('Tien'))

def add_ten(num):
    ten = 10
    return num + ten
print(add_ten(90))

def square_number(x):
    return x * x
print(square_number(2))

def area_of_circle (r):
    PI = 3.14
    area = PI * r ** 2
    return area
print(area_of_circle(10))

def sum_of_numbers(n):
    total = 0
    for i in range(n+1):
        total+=i
    print(total)    
print(sum_of_numbers(10)) 
```

```python
# Hàm có tham số và giá trị trả về
def calculate_area(length, width):
    """Tính diện tích hình chữ nhật"""
    area = length * width
    return area

# Sử dụng hàm
area = calculate_area(5, 3)
print(f"Diện tích: {area}")
```
